```python
import matplotlib.pyplot as plt

# Generate the dataset
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(age, height, s=weight, alpha=0.5, c='blue', label='People')
plt.title('Healthcare Dataset - Age vs Height with Weight')
plt.xlabel('Age')
plt.ylabel('Height')
plt.legend()
plt.grid(True)
plt.show()

# Bar Chart
plt.figure(figsize=(8, 6))
plt.bar(people, weight, color='green')
plt.title('Healthcare Dataset - People vs Weight')
plt.xlabel('People')
plt.ylabel('Weight')
plt.grid(True)
plt.show()

# Histogram
plt.figure(figsize=(8, 6))
plt.hist(age, bins=5, color='orange', edgecolor='black', alpha=0.7)
plt.title('Healthcare Dataset - Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
